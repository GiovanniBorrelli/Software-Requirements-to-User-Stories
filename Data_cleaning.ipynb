```python
import pandas as pd

# Carica il CSV (sostituisci 'percorso/del/file.csv' con il path reale)
df = pd.read_csv('reqs_utf8.csv')

# Mostra informazioni iniziali
print("Numero di colonne iniziali:", df.shape[1])
print("Numero di righe iniziali:", df.shape[0])
print("Numero di valori nulli iniziali nella colonna 'sentence':", df['sentence'].isnull().sum())
print("Numero di duplicati iniziali nella colonna 'sentence':", df['sentence'].duplicated().sum())

# Rimuovi righe con 'sentence' nullo
df_clean = df.dropna(subset=['sentence'])

# Rimuovi righe duplicate basate sulla colonna 'sentence'
df_clean = df_clean.drop_duplicates(subset=['sentence'])

# Mostra informazioni dopo la pulizia
print("\n--- Dopo la pulizia ---")
print("Numero di righe rimanenti:", df_clean.shape[0])
print("Numero di valori nulli nella colonna 'sentence':", df_clean['sentence'].isnull().sum())
print("Numero di duplicati nella colonna 'sentence':", df_clean['sentence'].duplicated().sum())

# Salva il dataset pulito (opzionale)
# df_clean.to_csv('dataset_pulito.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carica il CSV
df = pd.read_csv('reqs_utf8.csv')

# Calcola valori
total_rows = df.shape[0]
duplicate_rows = df['sentence'].duplicated().sum()
null_rows = df['sentence'].isnull().sum()

# Dati e etichette
values = [total_rows, duplicate_rows, null_rows]
labels = ['Totale righe', 'Duplicati', 'Nulli']

# Crea grafico
plt.figure(figsize=(6,5))
plt.bar(labels, values, color=['skyblue', 'salmon', 'lightgreen'])
plt.ylabel('Numero di righe')

# Mostra grafico
plt.show()
```
